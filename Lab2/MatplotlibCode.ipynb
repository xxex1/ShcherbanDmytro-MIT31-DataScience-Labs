```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
x = np.linspace(0, 10, 20) 
y_true = 2 * x + 1
```

```python
noise = np.random.normal(0, 2, size=x.shape) 
y_measured = y_true + noise
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, y_true, label="Істинна функція (y=2x+1)", color="blue")
plt.scatter(x, y_measured, label="Виміряні дані", color="red", marker="o")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Лінійна функція з похибками вимірювань")
plt.legend()
plt.grid(True)
plt.show()
```

```python
mae = mean_absolute_error(y_true, y_measured)
mse = mean_squared_error(y_true, y_measured)
```

```python
print("MAE (Mean Absolute Error):", mae)
print("MSE (Mean Squared Error):", mse)
```

```python
results = pd.DataFrame({
    "Metric": ["MAE", "MSE"],
    "Value": [mae, mse]
})
results.to_csv("error_metrics.csv", index=False)

print("Результати збережено у 'error_metrics.csv'")
```
