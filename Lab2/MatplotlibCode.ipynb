```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
np.random.seed(42)
X = np.linspace(0, 10, 50) 
true_slope = 2.0
true_intercept = 1.0
Y_true = true_slope * X + true_intercept
```

```python
noise = np.random.normal(0, 2, size=X.shape)
Y = Y_true + noise
```

```python
Y_hat = Y_true
```

```python
mae = mean_absolute_error(Y, Y_hat)
mse = mean_squared_error(Y, Y_hat)

print("MAE:", mae)
print("MSE:", mse)
```

```python
plt.scatter(X, Y, label="Виміряні дані (з шумом)", color="blue", alpha=0.6)
plt.plot(X, Y_hat, label="Ідеальна лінія (Y_hat)", color="red", linewidth=2)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Лінійна функція з шумом")
plt.legend()
plt.grid(True)
plt.show()
```

```python
df = pd.DataFrame({
    "X": X,
    "Y": Y,
    "Y_hat": Y_hat
})
df["mAE"] = mae
df["mSE"] = mse

output_path = "errors.csv"
df.to_csv(output_path, index=False)
```
