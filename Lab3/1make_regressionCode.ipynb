```python
import pandas as pd
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt
```

```python
X, y = make_regression(
    n_samples=100,       # Кількість зразків
    n_features=1,        # Кількість ознак
    noise=10,            # Додавання шуму
    random_state=42      # Фіксований генератор для відтворюваності
)
```

```python
df = pd.DataFrame({'X': X.flatten(), 'y': y})
print(df.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['X']], df['y'], test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print(f'MSE: {mse:.2f}, MAE: {mae:.2f}')
```

```python
plt.figure(figsize=(6,4))
plt.scatter(X_test, y_test, color='blue', label='Реальні дані')
plt.plot(X_test, y_pred, color='red', label='Прогноз')
plt.title('Регресія для штучних даних (make_regression)')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
