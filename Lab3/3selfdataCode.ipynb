```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
X_np = np.linspace(0, 50, 150)
```

```python
noise = np.random.normal(0, 8, size=X_np.shape) 
y_np = 2.5 * X_np + noise
```

```python
df_np = pd.DataFrame({'X': X_np, 'y': y_np})
print(df_np.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_np[['X']], df_np['y'], test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print(f'MSE: {mse:.2f}, MAE: {mae:.2f}')
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_test, y_test, color='blue', alpha=0.5, label='Реальні дані')
plt.plot(X_test, y_pred, color='red', label='Прогноз')
plt.title('Регресія для даних з Numpy')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
